```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# from google.colab import data_table
# data_table.enable_dataframe_formatter()

print("Libraries loaded as expected")
```

```python
df = pd.read_csv("Planetary_data.csv", comment="#")
df.head()
```

```python
# df.info() # helps understand basic structure of the dataset.
# df.value_counts() # helps understand unique value counts.
```

```python
# planet discovery distribution - which planet was discovered in which year

# disc_year_dist = df.groupby("disc_year")['pl_name'].first()
# print(disc_year_dist)
```

```python
# largest and smallest planet distribution.

## largest and smallest planet based on Planet Mass or Mass*sin(i) [Earth Mass]
print('largest planet by planet mass')
massive_planets_by_planet_mass = df[['pl_name', 'disc_year', 'pl_bmasse']].sort_values('pl_bmasse', ascending=False).head(20)
print(massive_planets_by_planet_mass)
```

```python
## smallest planet by planet mass
print('smallest planet by planet mass')
smallest_planets_by_planet_mass = df[['pl_name', 'disc_year', 'pl_bmasse']].sort_values('pl_bmasse').head(20)
print(smallest_planets_by_planet_mass)
```

```python
# planets and their orbital time

smallest_orbit_time_planets = df[['pl_name', 'pl_orbper']].sort_values('pl_orbper').head(20)
print(smallest_orbit_time_planets)
```

```python
largest_orbit_time_planets = df[['pl_name', 'pl_orbper']].sort_values('pl_orbper', ascending=False).head(20)
print(largest_orbit_time_planets)
```

```python
# planets with mass greater than Earth mass

df[df['pl_bmasse'] > 10 ]
```

```python
# recent discoveries

most_recent_discoveries = df[['pl_name' ,'disc_year']].sort_values(by='disc_year',ascending=False).head(30)
print(most_recent_discoveries)
```

```python
# Exploring exoplanets near our solar system.

## Exoplanet hunting evolution over time.

exoplant_search_trend = df.groupby(['disc_year', 'discoverymethod']).size().head(20)
print(exoplant_search_trend)
```

```python
# Who are the most prolific planet hunters?
## How many planets did each facility discover per year?

df.groupby(['disc_facility', 'disc_year']).size().head(20)
```

```python
# How many planets did each facility discover overall?
exoplanets_facitlity_dist = df.groupby('disc_facility').size().sort_values(ascending=False).head(10)
print(exoplanets_facitlity_dist)
```

```python
# Most effective discovery method.

effect_disc_method = df.groupby('discoverymethod').size().sort_values(ascending=False)
print(effect_disc_method)
```

```python
# efficient discovery methods
efficient_disc_method = df['discoverymethod'].value_counts().sort_values(ascending=False)
print(efficient_disc_method)
```

```python
## discovery method and disc_facility relation.

disc_method_and_facility = df[['discoverymethod', 'disc_facility']].value_counts().sort_values(ascending=False).head(15)
print(disc_method_and_facility)
```

```python
## select columns first and then perform aggregation.

### Methods and type of planets they find.
# Mass (Earth masses)	Interpretation
# < 10	              Earth-like / small planets
# 10–100	            Neptune-like
# 100–1000	          Gas giants
# 1000+	              Very massive gas giants / super-Jovian

method_planet_dist = df.groupby('discoverymethod')[['pl_bmasse', 'pl_rade']].agg(['median'])
print(method_planet_dist)
```

```python
## Are planets discovered closer to Earth (smaller distance sy_dist) more likely to be Earth-sized?
#### 1 is Earth radii. Jupiter is 11.2 radii. Neptune = 3.9 approx.
#### Since detecting small planets over long distances is difficult. Data starts skewing to large planets.

df['dist_category']  = np.where(df['sy_dist'] < 100, "Close", "Far")
df.groupby('dist_category')['pl_rade'].agg(['mean', 'median'])
```

```python
## Explore relation between orbital time and planetary mass. No corelation exists between planet mass and orbit.

df[['pl_bmasse', 'pl_orbper']].corr(method='spearman')
```

```python
## disc_facility and their discovery range.
disc_facility_range = df.groupby('disc_facility')['disc_year'].agg(['max','median','min']).head(20)
print(disc_facility_range)
```

```python
## Habitable zone i.e. goldilocks zone
### Earth-like conditions.

df.loc[
    (df['pl_eqt'].between(273,373, inclusive='neither')) & (df['pl_rade'] < 2.0) ,
    ['pl_name', 'hostname']
].head(25)
```

```python
df['sun_cat'] = np.where(df['st_teff'] < 5000, "Red Dwarfs", "Sun like planets" )
counts = df.groupby('sun_cat')['pl_name'].count()
print(counts)
```

```python
## Compare star temperature to planet equilibrium temparature.

df.plot(
    kind="scatter",
    x='st_teff',
    y='pl_eqt'
)
plt.title("Star temperature vs planet temperature distribution")
plt.xlabel("Star-temp")
plt.ylabel("Planet-temp")

plt.show()
```

```python
## visualize planet temperature distribution.

df['pl_eqt'].plot(kind='hist', bins=30)
plt.show()
```

```python
## Exploring orbital periods of exoplanets.

# earth takes one year or 365 days to orbit the sun. In astronomy it's called orbital period.

## planets with shortest orbital periods

shortest_orbital_periods = df.groupby('pl_orbper')[['pl_name']].agg(['min']).head(10)
print(shortest_orbital_periods)
```

```python
## longest orbital periods.

longest_orbital_periods = df[['pl_orbper', 'pl_name']].sort_values(by='pl_orbper', ascending=False).head(10)
print(longest_orbital_periods)
```

```python
## how long would it take for me to celebrate a new year on coconuts-2 b the largest exoplanet on the basis of orbital periods.
### ~1.1 million years between two years.
#### How many lifetimes?
#### average earth life time ~71 years https://ourworldindata.org/life-expectancy

longest_orbital = df[['pl_orbper', 'pl_name']].sort_values(by='pl_orbper', ascending=False).head(1)
next_new_year = longest_orbital['pl_orbper'] / 365

life_times = next_new_year / 71
print(f"It would take {life_times} before the next new year on the planet {longest_orbital['pl_name'].iloc[0]}")
```

```python
## analysing exo_planet year

exo_planey_yr = df[['pl_orbper']].agg(['median'])
print(f"The median exoplanet year {exo_planey_yr.iloc[0]} is closer to mercury (88 days) than Earth (365 days)")
```

```python
## Distribution of planets with orbital period less than 365

short_years = df[df['pl_orbper' ] < 365 ]
short_years['pl_orbper'].plot(kind='hist', bins=50, title="Distribution of Exoplanet Years (Under 365 Days)")
```
