```python
# largest and smallest planet distribution.

## largest and smallest planet based on Planet Mass or Mass*sin(i) [Earth Mass]
print('largest planet by planet mass')
massive_planets_by_planet_mass = df[['pl_name', 'disc_year', 'pl_bmasse']].sort_values('pl_bmasse', ascending=False).head(20)
print(massive_planets_by_planet_mass)
```

```python
## smallest planet by planet mass
print('smallest planet by planet mass')
smallest_planets_by_planet_mass = df[['pl_name', 'disc_year', 'pl_bmasse']].sort_values('pl_bmasse').head(20)
print(smallest_planets_by_planet_mass)
```

```python
# planets and their orbital time

smallest_orbit_time_planets = df[['pl_name', 'pl_orbper']].sort_values('pl_orbper').head(20)
print(smallest_orbit_time_planets)
```

```python
largest_orbit_time_planets = df[['pl_name', 'pl_orbper']].sort_values('pl_orbper', ascending=False).head(20)
print(largest_orbit_time_planets)
```
